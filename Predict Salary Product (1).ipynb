```python
##graph
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#ukuran
plt.figure(figsize=(16,6))

#scatter plot graph
sns.scatterplot(x=salary_data['YearsExperience'], y=salary_data['Salary'])
plt.title("Predict Salary Product")

```

```python

#ukuran
plt.figure(figsize=(16,6))

#scatter plot graph
sns.lineplot(data=salary_data)
plt.xlabel(' YearsExperience')
plt.ylabel('Price')
plt.title("Predict Salary Product")
```

```python
#machine learning model
#import model liner regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#corelation
salary_data.corr()
```

```python
#menentukan variable x dan y
x = salary_data['YearsExperience'].values.reshape(-1,1)
y = salary_data['Salary'].values.reshape(-1,1)

#melatih data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

#metode linear
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)

#menetukan a,b awalan rumus y=ax-b
a=lin_reg.coef_
b= lin_reg.intercept_

print(a)
print(b)

```

```python
#akurasi model
acurracy = lin_reg.score(x_test, y_test)
print(acurracy)
```

```python
#gambar prediksi

plt.figure(figsize=(11,6))
y_prediksi = lin_reg.predict(x_test)
plt.scatter(x_test, y_test)
plt.plot(x_test, y_prediksi, c='r')
plt.xlabel(' YearsExperience')
plt.ylabel('Price')
plt.title("Predict Salary Product")
```
